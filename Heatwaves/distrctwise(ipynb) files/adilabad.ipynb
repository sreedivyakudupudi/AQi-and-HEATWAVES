```python
month=input("enter the month you want to predict in 3-letter format(Jan): ")
```

```python
df1=pd.DataFrame()
for index, row in df.iterrows():
    if month in row['Row Labels'] :
        df1 = df1.append(row)
print(df1)
```

```python
df1["target"]=df1.shift(-1)["Average of temp_max (â°C)"]
df1
```

```python
df2=df1.tail(1)
df2.drop(['Row Labels','target'],axis=1,inplace=True)
```

```python
df2
```

```python
x=np.array(df2)
x
```

```python
df1=df1[df1.target.notnull()]
df1
```

```python
predictors = df1.columns[~df1.columns.isin(["target", "Row Labels"])]
df1[predictors].head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X=df1[predictors]
y=df1["target"]
X
```

```python
X=sc.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 300, max_features = 'sqrt', max_depth = 5, random_state = 18)
```

```python
rf.fit(X,y)
y_pred=rf.predict(X)
y_pred.shape
```

```python
print("predicting the temperatures of previous years: ",y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
error=mean_absolute_percentage_error(y_pred,y)
```

```python
print("mean absolute percentage error for ",month," is: ",error)
```

As Adilabad is a hilly region there are chances for occurance of heatwaves if the maximum temperature exceeds 35 degree celsius
